```python
from pathlib import Path

import pandas as pd

merchant_kb_path = Path("merchant_kb.csv")
latest_n = 20

df = pd.read_csv(merchant_kb_path)
df["category_updated_at"] = pd.to_datetime(df["category_updated_at"], errors="coerce")

updated_rows = (
    df[df["category_updated_at"].notna()]
    .sort_values("category_updated_at", ascending=False)
    .head(latest_n)
)

print(f"recently updated merchants: {len(updated_rows)}")
display(updated_rows)
```

```python
from pathlib import Path

import pandas as pd

merchant_kb_path = Path("merchant_kb.csv")
output_path = Path("merchant_names.csv")

# 读取原始 CSV
df = pd.read_csv(merchant_kb_path)

# 只保留 merchant_name 列
merchant_names = df[["merchant_name"]]

# 保存为新的 CSV，不修改原文件
merchant_names.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"total merchants: {len(merchant_names)}")
print(f"saved to: {output_path.resolve()}")

display(merchant_names)
```
